```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_marketing_funnel_data.csv")
```

```python
channel_analysis = df.groupby("channel").agg({
    "visited_website": "sum",
    "purchase_completed": "sum",
    "revenue": "sum"
})
```

```python
channel_analysis["conversion_rate"] = (
    channel_analysis["purchase_completed"] /
    channel_analysis["visited_website"]
) * 100
```

```python
channel_analysis.sort_values("conversion_rate", ascending=False)
```

```python
channel_analysis["conversion_rate"].sort_values().plot(
    kind="barh",
    figsize=(8,5)
)

plt.title("Conversion Rate by Channel")
plt.xlabel("Conversion Rate (%)")
plt.show()
```

```python
campaign_analysis = df.groupby("campaign_type").agg({
    "visited_website": "sum",
    "purchase_completed": "sum",
    "revenue": "sum"
})

campaign_analysis["conversion_rate"] = (
    campaign_analysis["purchase_completed"] /
    campaign_analysis["visited_website"]
) * 100

campaign_analysis.sort_values("conversion_rate", ascending=False)
```
