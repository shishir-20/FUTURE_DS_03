```python
import pandas as pd

df = pd.read_csv("../data/raw/d2c_marketing_funnel_data.csv")
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
yes_no_cols = [
    "visited_website",
    "viewed_product",
    "added_to_cart",
    "checkout_started",
    "purchase_completed",
    "discount_applied"
]

for col in yes_no_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
df["purchase_completed"].sum()
```

```python
df.to_csv("../data/processed/cleaned_marketing_funnel_data.csv", index=False)
```
