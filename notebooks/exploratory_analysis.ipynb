```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_marketing_funnel_data.csv")
```

```python
print("Total Sessions:", df.shape[0])
print("Total Revenue:", df["revenue"].sum())
print("Average Order Value:", df["order_value"].mean())
print("Total Purchases:", df["purchase_completed"].sum())
```

```python
df["channel"].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="channel")
plt.title("Traffic by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df["campaign_type"].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="campaign_type")
plt.title("Campaign Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="device")
plt.title("Traffic by Device")
plt.show()
```
